# 511 Transit Data - API calls and responses

```python
import os
import sys
import json
import requests
import pandas as pd
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
API_KEY = os.getenv('MUNI_API_KEY')
# API_URL = "http://api.511.org/transit/gtfsoperators?api_key={}".format(API_KEY)
```

```python
API_URL = os.getenv('API_URL') + API_KEY
```

```python
response = requests.get(API_URL)
response.status_code
```

```python
if response.status_code == 200:
    response_bytes = response.content
    response_decoded_string = response_bytes.decode('utf-8-sig')
    data = json.loads(response_decoded_string)
```

```python
data
```

```python
df = pd.DataFrame(data)
df = df.rename(columns = {"Id": "system_id", "Name": "system_name", "LastGenerated": "last_generated"}, errors = "raise")
print(df.shape)
```
